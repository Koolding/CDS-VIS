Then I want to make a color histogram which will be used for comparison to the other images

```python
# split channels
channels = cv2.split(image)
# names of colours
color_names = ["b", "g", "r"]
# create figure
plt.figure()
# add title
plt.title("Multiple channels")
# add xlabel
plt.xlabel("Bins")
# add ylabel
plt.ylabel("Number of pixels")

# loop over our channels
for channel, color_name in zip(channels, color_names):
    # Create a histogram
    hist = cv2.calcHist([channel], [0], None, [256], [0,256])
    # Plot histogram
    plt.plot(hist, color = color_name)
    # Set limit of x-axis
    plt.xlim([0, 256])

# show the plot
plt.show()
```
